```python
# Gráfico da relação - Nível de experiência|Salário em dólar:

sns.set_style("darkgrid")
figura, graficos = plt.subplots(1, 1, sharey=True, figsize=(4,6))
plt.tight_layout()

sns.boxplot(data=df_ds, y='salary_in_usd', x='experience_level')
```

```python
# Criando um Dataframe com a relação - Tipo de contrato|Salário em dólar:

df_contrato = df_ds.loc[:, ['employment_type'] + ['salary_in_usd']]
df_contrato.head()
```

```python
# Gráfico da relação - Tipo de contrato|Salário em dólar:

sns.set_style("darkgrid")
figura, graficos = plt.subplots(1, 1, sharey=True, figsize=(4,6))
plt.tight_layout()

sns.boxplot(data=df_ds, y='salary_in_usd', x='employment_type')
```

```python
# Criando um Dataframe com a relação - Tipo de trabalho|Salário em dólar:

df_contrato = df_ds.loc[:, ['remote_ratio'] + ['salary_in_usd']]
df_contrato.head()
```

```python
# Gráfico da relação - Tipo de trabalho|Salário em dólar:

sns.set_style("darkgrid")
figura, graficos = plt.subplots(1, 1, sharey=True, figsize=(4,6))
plt.tight_layout()

sns.boxplot(data=df_ds, y='salary_in_usd', x='remote_ratio')
```

```python
# Criando um Dataframe com a relação - Porte da empresa|Salário em dólar:

df_contrato = df_ds.loc[:, ['company_size'] + ['salary_in_usd']]
df_contrato.head()
```

```python
# Gráfico da relação - Porte da empresa|Salário em dólar:

sns.set_style("darkgrid")
figura, graficos = plt.subplots(1, 1, sharey=True, figsize=(4,6))
plt.xticks(rotation = 0)
plt.tight_layout()

sns.boxplot(data=df_ds, y='salary_in_usd', x='company_size')
```

```python
# Visualizando graficamente os salários médio em dólar de cada função:

plt.subplots(figsize=(10,10))
sns.barplot(y='job_title',x='salary_in_usd', data = df_ds)
```

```python
# Cargos melhores remunerados mundialmente e os níveis de experiência:

df_ds[['job_title','experience_level','salary_in_usd']].sort_values(by='salary_in_usd',ascending=False).drop_duplicates('job_title').reset_index(drop=True).head(10)
```

# Conclusão

Depois de realizar uma análise detalhada dos dados obtidos, com o objetivo de avaliar as possíveis relações entre características dos colaborador e os salários em dólar dos mesmos, podemos afirmar que existem algumas evidências de correlações entre os salários em dolar:

- Ano trabalhado (principalmente 2022)
- Experiência do colaborador (Executivo e Sênior)
- Tipo do contrato (PJ e Tempo integral)
- Tipo de trabalho (Home-Office)
- Porte da empresa que o colaborador trabalha (Médio e Grande)

Podemos confirmar com essa amostra que o ano de 2022, profissional nível executivo, tipo de contrato PJ, estilo de trabalho Home-Office e empresas de médio porte tem relação com os salários mais altos em dólar.

O cientista de dados vem ganhando muita notoriedade, por estar dentro dos 10 cargos que são melhores remunerados, Data Science ocupa a 6 posição. Porém os profissionais com maior nível de experiência ainda tem os salarios mais altos.
